```python
import os
from data import load_data
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
_, _, aug_test_data = load_data('mnist', binarize = False, augment = True, val = False)
_, _, test_data = load_data('mnist', binarize = False, augment = False, val = False)
aug_test_loader = DataLoader(aug_test_data, batch_size=2, shuffle=False, num_workers=os.cpu_count())
test_loader = DataLoader(test_data, batch_size=2, shuffle=False, num_workers=os.cpu_count())
#aug_test_val, aug_test_bpd = evaluation(aug_test_loader, device, model_type = model_type, model_best=model)
```

```python
img1 = next(iter(test_loader))[0].view(28,28)
```

```python
img2 = next(iter(aug_test_loader))[0].view(28,28)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axes[0].imshow(img1, cmap = 'gray')
axes[0].axis('off')  # optional: hide axis ticks and labels
axes[1].imshow(img2, cmap = 'gray')
axes[1].axis('off')
plt.tight_layout()
plt.show()
plt.axis('off')
```
